```python
import numpy as np

# we can populate an array with regularly-spaced numbers (notice how the endpoints are INCLUDED)
x = np.linspace(1, 10, num=10)
x
```

```python
# since we're using integers to populate we also could've done this (the end value is EXCLUDED)
np.arange(1, 11)
```

```python
x = np.linspace(-np.pi, np.pi, num=50)
x
```

```python
# we can plot 3d stuff using contour plots
import matplotlib.pyplot as plt
%matplotlib inline

# let's create a regular square grid
y = x
X, Y = np.meshgrid(x, y)

# this will compute Z values
def f(x, y):    
    return np.cos(y)/(1-x**2)


fig = plt.figure(1)

ax = fig.add_axes([0.0, 0.0, 1.0, 1.0], frameon=True)
ax.contour(X, Y, f(X,Y), 15)  # use 15 contours

ax.set_title("Contour Plot")
ax.set_xlabel("this is the x-axis")
ax.set_ylabel("this is the y-axis")

plt.show()
```

```python
# here's the same thing, but a heatmap instead of contour lines
fig = plt.figure(1)

ax = fig.add_axes([0.0, 0.0, 1.0, 1.0], frameon=True)
ax.contourf(X, Y, f(X,Y), 15)

ax.set_title("Heatmap Plot")
ax.set_xlabel("this is the x-axis")
ax.set_ylabel("this is the y-axis")

plt.show()
```

```python
# we can also just plot the surface in 3d
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(1)

ax = fig.add_axes([0.0, 0.0, 1.0, 1.0], projection='3d')  # this creates an Axes3D object
ax.plot_surface(X, Y, f(X,Y))

ax.set_title("3D Plot")
ax.set_xlabel("this is the x-axis")
ax.set_ylabel("this is the y-axis")
ax.view_init(elev=30, azim=40)  # set the viewing angle

plt.show()
```

# Indexing Data

```python
import numpy as np

A = np.arange(1, 17)
A = A.reshape((4,4), order='F')
A
```

```python
# we can get the value at a single index
A[1,2]  # although in R we have 1-based indexing, in python we have 0-based indexing
```

```python
# or we can grab multiple values by specifying a list for each dimension 
# (so this grabs the values at (0,1) and (2,3))
A[[0,2], [1,3]]
```

```python
# we can also grab values using all possible combinations of indices (i.e. (0,1), (0,3), (2,1), (2,3))
A[np.ix_([0,2], [1,3])]
```

```python
# we can grab ranges (again, the end of the range is EXCLUDED)
A[0:3, 1:4]
```

```python
# if we don't specify ranges in a specific dimension then we get the full range
A[0:2,]

# this is the same as A[0:2,:]
```

```python
# ranges and indices can be mixed
A[0:2,[0,3]]
```

```python
# we can't have empty before the comma, so we always use a colon : to indicate "full range"
A[:,0:2]
```

```python
# there is also a powerful "boolean mask" method of indexing

mask = np.zeros(A.shape, dtype=np.bool)  # start by masking everything out
mask[:,0:2] = 1  # add back in the indices that we want to keep
A[mask]
```

```python
# notice how the block structure was lost using the boolean mask method of indexing.  We restore the shape manually
A[mask].reshape((4,2))
```

```python
# there is no such thing as "negative indices" (like in R)
# however, we can simulate it by using boolean masks

mask = np.ones(A.shape, dtype=np.bool)  # start by keeping everything
mask[[0,2],:] = 0  # now throw away the 0th and 2nd row
A[mask]
```

```python
# again, the shape is lost.  We set it manually
A[mask].reshape((2,4))
```

```python
# as you have already guessed, we can always get the dimensions of the array using the "shape" property
A.shape
```

# Loading Data

```python
# What if we have heterogeneous data in a table? Like this:

# mpg     cylinders       displacement    horsepower weight       acceleration    year    origin  name
# 18.0   8   307.0      130.0      3504.      12.0   70  1        "chevrolet chevelle malibu"
# 15.0   8   350.0      165.0      3693.      11.5   70  1        "buick skylark 320"
# 18.0   8   318.0      150.0      3436.      11.0   70  1        "plymouth satellite"
# ...

# We *could* use a numpy array where dtype=object (a compound datastructure to hold each row).
#
# Or we could maintain 9 numpy arrays side-by-side (one for mpg, one for cylinders, etc) and handle the index 
# bookkeeping to make sure that our records stay in sync.

# There's a powerful library that will hide this nastiness from us
import pandas as pd
from IPython.display import display

df = pd.read_csv("Auto.data", delimiter=r"\s+", na_values=['?'])  # read whitespace-delimited file, interpret ? as missing data
display(df.head(5))
```

```python
# Reading a plain CSV version of the data is almost the same

# mpg,cylinders,displacement,horsepower,weight,acceleration,year,origin,name
# 18,8,307,130,3504,12,70,1,chevrolet chevelle malibu
# 15,8,350,165,3693,11.5,70,1,buick skylark 320
# ...

df = pd.read_csv("Auto.csv", na_values=['?'])  # read comma-delimited file, interpret ? as missing data
display(df.head(5))
```

```python
# how many rows are there?
len(df)
```

```python
# we can drop any rows that have missing values
df = df.dropna()
len(df)
```

```python
# display the names of the columns
df.columns.values
```

# Additional Graphical and Numerical Summaries

```python
# It's easy to create a scatterplot of one column vs another
# The notation df['mpg'].values fetches the raw underlying numpy array
import pandas as pd
import numpy as np

df = pd.read_csv("Auto.csv", na_values=['?'])

import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(1)

ax = fig.add_axes([0.0, 0.0, 1.0, 1.0], frameon=True)
ax.plot(df['cylinders'].values, df['mpg'].values, 'o')

ax.set_title("Plot of Cylinders vs MPG")
ax.set_xlabel("Cylinders")
ax.set_ylabel("MPG")

plt.show()
```

```python
# We want to reinterpret the "cylinders" column as a factor/categorical type variable
df['cylinders'] = df['cylinders'].astype(object)
```

```python
# Since we have a factor variable, we may want boxplots instead.  Pandas has a convenience wrapper around matplotlib
df.boxplot(column='mpg', by='cylinders')
```

```python
# Since pandas.DataFrame.boxplot is just a convenience wrapper around matplotlib.pyplot.boxplot, I can pass extra arguments and
# they just get forwarded to the underlying boxplot fn
df.boxplot(column='mpg', by='cylinders', vert=False)
```

```python
# Histograms are easy
fig = plt.figure(1)

ax = fig.add_axes([0.0, 0.0, 1.0, 1.0], frameon=True)
ax.hist(df['mpg'].values, bins=10)

ax.set_title("Histogram of MPG")
ax.set_xlabel("MPG")
ax.set_ylabel("Frequency")

plt.show()
```

```python
# it is also easy to plot a scatter matrix (pairwise scatter plot)
from pandas.tools.plotting import scatter_matrix

scatter_matrix(df, figsize=(12,12), diagonal='kde')
```

```python
# We can easily leave some columns out of our scatter matrix
brief_df = pd.DataFrame(df, columns=['mpg', 'displacement', 'horsepower', 'weight', 'acceleration'])
display(brief_df.head(5))
```

```python
scatter_matrix(brief_df, figsize=(12,12), diagonal='kde')
```
